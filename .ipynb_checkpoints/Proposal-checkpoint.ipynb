# API Project Proposal: Bus Reliability Analysis Across Major Northeast Transit Agencies

## Question:
What route and temporal factors are associated with bus bunching on high-frequency corridors, and do these relationships vary across different transit agencies (NYC, Boston, and DC)?

## Background & Movtivation:
Bus bunching is very disruptive failures in urban transit systems, it happens when two buses on the same route arrive nearly simultaneously after a long gap which leads to doubling or even tripling passenger wait times without any change in the number of buses operating. Bunching means that wait time is unpredictable and often far longer than it should be, which disproportionately affects riders with no alternative to the bus.

Transit agencies track on-time performance, but on high frequency routes, schedule adherence is the wrong metric. A bus can be perfectly on time and still cause bunching if the bus behind it is also on time. What matters on these corridors is headway regularity, whether the gaps between buses stay consistent and that is measured far less often and understood far less clearly than schedule adherence.

What makes this interesting to me is that I'm not just measuring bunching, but measuring it comparatively across three cities. NYC, Boston, and DC share broadly similar characteristics such as dense urban environments, high transit dependency, and large bus networks that are affected traffic, but they differ in infrastructure investment, operations, and route design. If bunching were purely a function of traffic and demand, we would expect similar patterns across all three. If differences show up across agencies even after accounting for route type and time of day, that means some systems are just run better than others, and that there's something agencies can learn and o better from. So that requries data across multiple systems which is what I hope to collect.

##### What is GTFS Data?
Real-time data will be collected from the GTFS-RT and Vehicle Positions feeds published by all three agencies. These feeds provide predicted arrival times by stop and trip, route IDs, vehicle GPS positions, and timestamps. GTFS-RT is a standardized open format developed by Google and now maintained as an industry standard across hundreds of transit agencies worldwide. Because all three agencies publish their real-time data in the same format, the data collection code can follow a consistent structure across all three rather than writing entirely separate logic for each agency. This means the variables constructed from each agency's feed are directly comparable by design, which is essential for a cross agency analysis.

##### Static GTFS Feed
Scheduled headway data will be derived from each agency's static GTFS feed, specifically the stop_times.txt file, which contains the scheduled arrival time for every trip at every stop. This is download and not in the  API, but is needed for a  baseline to measure against since headway deviation and the bunching indicator cannot be constructed without knowing what the scheduled gap was supposed to be in the first place. All three agencies publish their static GTFS feeds and they are updated regularly to reflect schedule changes.

## Variables:

##### Variable 1: Agency Dummies (binary)

Two binary variables — one equal to 1 if the observation is from MTA and another one equal to 1 if it's from WMATA, with MBTA serving as the reference category. 

##### Variable 2: Headway Deviation (Continuous)

The difference in minutes between the actual observed gap between consecutive buses and the scheduled gap. Constructed by comparing real-time arrival predictions across vehicle observations on the same route, stop, and direction. 

##### Variable 3: Bunching Indicator (Binary) 

A binary flag equaling 1 when the actual headway falls below 50% of the scheduled headway, indicating two buses are running close together. This is the primary unique variable. It does not exist in any API response or open data portal, it'll be constructed by comparing consecutive vehicle arrival predictions. 

##### Variable 4: Scheduled Headway (Continuous)

The planned gap in minutes between buses, derived from each agency's GTFS static schedule feed and matched to real-time observations by route, stop, and time of day.

##### Variable 5: Peak Hour Indicator (Binary)

A binary variable equal to 1 during morning peak (6–9am) and evening peak (4–7pm), and 0 otherwise. Constructed  from the timestamps recorded during each API poll. 

##### Variable 6: Schedule Adherence Ratio (Continuous)

Actual headway divided by scheduled headway. A ratio of 1.0 means buses are spaced exactly as planned, below 1.0 means buses are closer together than scheduled, above 1.0 means the gap is larger than scheduled. Unlike raw headway deviation which is in minutes, this ratio normalizes across routes with different scheduled frequencies which will make it a fair basis for cross-agency comparison since MBTA, MTA, and WMATA routes don't all run on the same schedule.

##### Variable 7: Headway Coefficient of Variation (Continuous)

Standard deviation of observed headways divided by mean headway, calculated per route per time period. The headway deviation will capture how far off any  observation is from schedule, the CV captures how consistent a route's service is across many observations. A route could have a moderate average deviation but extremely high variance which would mean service is sometimes fine and sometimes terrible.
